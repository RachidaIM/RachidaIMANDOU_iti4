# Filtrage

```python
#la creation du masque de convolution pour le filtrage gaussien
Hg = np.array([[1,2,1],[2,4,2], [1,2,1]])/16
```

```python
def filtrGaussien(img):
    #convolution entre la l'image et la matrice H
    B = signal.convolve2d(img, Hg, boundary='fill', mode='same')
    plt.imshow(B,'gray')
```

```python
#affichage de l'image filtree
plt.figure()
filtrGaussien(tumeur)
plt.title('image filtree')
plt.figure()
plt.imshow(tumeur,'gray')
plt.title('image original')
```

## Bruitage, filtrage et Détection de contoure

```python
def seuillage(I,seuil=None):
    uint8_I = (255*((I-I.min())/(I.max()-I.min()))).astype(np.uint8)
    return uint8_I>seuil
```

```python
def edgeExtractor(img,methode,seuil=None):
    
    # Extraction des contours
    contours_img,__ = methode(img)
    
    # Seuillage des contours
    if seuil is not None:
        thresh_img = seuillage(contours_img,seuil)
    else:
        # Si le seuil n'est pas défini, on va ici faire une binarisation avec seuillage automatique (Otsu)
        uint8_img = (255*((contours_img-contours_img.min())/(contours_img.max()-contours_img.min()))).astype(np.uint8)
        seuil,thresh_img = cv2.threshold(uint8_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        print("Seuillage Otsu : {}".format(seuil))
    return thresh_img
```

```python
def salt_and_pepper_noise(img,p):
    # Création masque aléatoire avec p% de 0 et (100-p)% de 1
    mask=(np.random.randint(100,size=img.shape)>p)
    # Image bruit avec aléatoirement du blanc et du noir
    new_img = np.random.randint(2,size=img.shape)*255
    # Tous les pixels de mask à 0, prennent la valeur de new_img (0 ou 255) dans l'image originale
    new_img[mask]=img[mask]
    return new_img.astype(np.uint8)

noisy_polype = salt_and_pepper_noise(tumeur,3)

plt.figure()
plt.subplots(2,1,figsize=(12,12))
plt.subplot(121)
plt.imshow(tumeur,'gray')
plt.title('Image à traiter du jour')
plt.subplot(122)
plt.imshow(noisy_polype,'gray')
plt.title('Image bruitée du jour')
plt.show()
```

```python
import itertools
from scipy.ndimage.interpolation import shift

def median_filter(img,n):
    return np.floor(np.nanmedian(np.stack([shift(img,(i-n//2,j-n//2),cval=np.NaN) for i,j in itertools.product(range(n),range(n))],axis=-1),axis=(-1))).astype(np.uint8)

preprocess_polype = median_filter(noisy_polype,3)

plt.figure()
plt.subplots(1,1,figsize=(6,6))
plt.imshow(preprocess_polype,'gray')
plt.title('Image filtrée ')
```

```python
def Laplacien(I):
    Laplacien=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]],dtype=np.float64)
    Lap_I=cv2.filter2D(I,-1,Laplacien)
    norme_I = np.linalg.norm([Lap_I],axis=0)
    return norme_I,Lap_I
```

```python
edge_polype = edgeExtractor(preprocess_polype,Laplacien)

plt.figure()
plt.subplots(1,1,figsize=(6,6))
plt.imshow(edge_polype,'gray')
plt.title('Image traitée du jour')
```

```python
def histogramme(image,titre):
    plt.hist(image.flatten(),bins= len(image.flatten()),histtype='step')
    plt.title(titre)
```

```python
#Telechargement de l'image
Melanome = cv2.imread(path_base+'D36_2_orig.jpg',0)
plt.imshow(Melanome,'gray')
```

```python
def convolution_faster(img,h):
    h_rows,h_cols=h.shape
    h = h.reshape((-1,))
    rows,cols = img.shape
    large_img = np.stack([shift(img.astype(np.float32),(i-h_rows//2,j-h_cols//2),cval=np.NaN) for i,j in itertools.product(range(h_rows),range(h_cols))],axis=-1)
    return np.dot(large_img,h).astype(np.uint8)
```

```python
expo_h = np.array([[1,1,3,1,1],
                  [1,3,7,3,1],
                  [3,7,16,7,3],
                  [1,3,7,3,1],
                  [1,1,3,1,1]],dtype=np.float64)/80
```

```python
#le filtrage exponentiel
Melanome_expo = convolution_faster(Melanome,expo_h)

plt.imshow(Melanome,'gray')
```

```python
histogramme(Melanome_expo,"histogramme image filtrée expo")
```

```python
#le filtrage gaussien
gauss_h = np.array([[1,2,1],[2,4,2],[1,2,1]],dtype=np.float64)/16

Melanome_fg = convolution_faster(Melanome,gauss_h)

plt.imshow(Melanome_fg,'gray')

```

```python
histogramme(Melanome_fg,"histogramme filtrée")
```

```python
#le filtrage moyenneur
def mean_filter_fast(img,n):
    return np.floor(sum([shift(img.astype(np.float32),(i-n//2,j-n//2),cval=np.NaN) for i,j in itertools.product(range(n),range(n))])/(n*n)).astype(np.uint8)

Melanome_moyenne = mean_filter_fast(Melanome,3)

plt.imshow(Melanome_moyenne,'gray')

```

```python
histogramme(Melanome_moyenne,"histogramme filtrée")
```

Maintenant nous allons passer au bruitage

```python
#Bruitage
noisy_Melanome = salt_and_pepper_noise(Melanome_bi,15)

plt.figure()
plt.subplots(2,1,figsize=(12,12))
plt.subplot(121)
plt.imshow(Melanome,'gray')
plt.title('Image à traiter du jour')
plt.subplot(122)
plt.imshow(noisy_Melanome,'gray')
plt.title('Image bruitée du jour')
plt.show()
```

```python
#Fitrage
Melanome_filtre = median_filter(noisy_Melanome,3)

plt.figure()
plt.imshow(Melanome_filtre,'gray')
plt.title('Image filtrée')
```

```python
#Detection de contour
def Laplacien(I):
    Laplacien=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]],dtype=np.float64)
    Lap_I=cv2.filter2D(I,-1,Laplacien)
    norme_I = np.linalg.norm([Lap_I],axis=0)
    return norme_I,Lap_I

def edgeExtractor(img,methode,seuil=None):
    
    # Extraction des contours
    contours_img,__ = methode(img)
    
    # Seuillage des contours
    if seuil is not None:
        thresh_img = seuillage(contours_img,seuil)
    else:
        # Si le seuil n'est pas défini, on va ici faire une binarisation avec seuillage automatique (Otsu)
        uint8_img = (255*((contours_img-contours_img.min())/(contours_img.max()-contours_img.min()))).astype(np.uint8)
        seuil,thresh_img = cv2.threshold(uint8_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        print("Seuillage Otsu : {}".format(seuil))
    return thresh_img

MelanomeP = edgeExtractor(Melanome_filtre,Laplacien)


output = cv2.cvtColor(Melanome_filtre,cv2.COLOR_GRAY2RGB)
output[MelanomeP==255]=[255,0,0]
plt.figure()
plt.subplots(1,1,figsize=(6,6))
plt.imshow(output,'gray')
plt.title('Résultat final')
```

```python
#comparaison
MelanomeVTT = cv2.imread(path_base+'D36_2_contour.png',0)
plt.figure()
plt.subplots(2,1,figsize=(12,12))
plt.subplot(121)
plt.imshow(output,'gray')
plt.title('Melanome Traité ')
plt.subplot(122)
plt.imshow(MelanomeVTT,'gray')
plt.title('Melanome  VTT ')
plt.show()
```

```python
#Erreur
def error(img1,img2):
    C = (img1.astype(np.int16)-img2.astype(np.int16)).astype(np.int32)
    return np.mean(np.abs(C))

error(Melanome_filtre,MelanomeVTT)
```

# Bonus 

Nous avons trouvé cette methode qui permet de detecter le tumeur dans l'image avec un rectangle

```python
image = cv2.imread(path_base+'D36_2_orig.jpg',-1)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lower_color = np.array([0, 50, 50])
upper_color = np.array([10, 255, 255])
mask = cv2.inRange(hsv, lower_color, upper_color)
contours, _= cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Afficher l'image
image = image[600:1400,:]
plt.imshow( image)
```
